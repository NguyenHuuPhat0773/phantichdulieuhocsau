### Ta cần tạo bảng phân phối tần số các từ trong copus, bảng này cần chuyển sang dạng list, ta dùng hàm list()

```python
all_words = nltk.FreqDist(w.lower() for w in reviews.words())
word_features = list(all_words)
```

### xác định một hàm để tính toán các đặc trưng, tức là những từ đủ quan trọng để thiết lập ý kiến của một review.

```python
def document_features(document, word_features):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['{}'.format(word)] = (word in document_words)
    return features
```

### Khi bạn định nghĩa hàm document_features(), bạn tạo 1 tập các documents

```python
featuresets = [(document_features(d,word_features), c) for (d,c) in documents]
len(featuresets)
```

```python
train_set, test_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

### Dùng thuật toán Naïve Bayes để phân loại, dùng thư viện NLTK. Sau đó tính toán độ chính xác của thuật toán

```python
train_set, est_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, test_set))
```

### 9.4 Viết chương trình Python với thư viện NLTK để loại bỏ các stopword từ một văn bản đã cho.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Tải dữ liệu cần thiết từ NLTK
nltk.download('punkt')
nltk.download('stopwords')

# Hàm loại bỏ stopwords từ văn bản
def remove_stopwords(text, language='english'):
    # Tokenize văn bản (chia văn bản thành các từ)
    words = word_tokenize(text)

    # Lấy danh sách stopwords cho ngôn ngữ đã chọn
    stop_words = set(stopwords.words(language))

    # Lọc bỏ các stopword
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Chuyển danh sách từ đã lọc thành chuỗi văn bản
    filtered_text = ' '.join(filtered_words)
    return filtered_text

# Văn bản mẫu
text = "This is a simple example to demonstrate how to remove stopwords from a text."

# Loại bỏ stopwords trong văn bản
filtered_text = remove_stopwords(text, language='english')

# In kết quả
print("Original Text: ", text)
print("Filtered Text: ", filtered_text)
```

### 9.5 Viết chương trình Python với thư viện NLTK bỏ qua các stopword từ danh sách các stopword.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def remove_stopwords(text):
    # Tải xuống danh sách stopwords nếu chưa có
    nltk.download('stopwords')
    nltk.download('punkt')

    # Lấy danh sách stopwords cho tiếng Anh
    stop_words = set(stopwords.words('english'))

    # Tokenize văn bản
    words = word_tokenize(text)

    # Lọc bỏ stopwords
    filtered_words = [word for word in words if word.lower() not in stop_words]

    return ' '.join(filtered_words)

# Văn bản đầu vào
text = "This is a simple example demonstrating how to remove stopwords using NLTK."

# Gọi hàm để loại bỏ stopwords
filtered_text = remove_stopwords(text)

print("Original Text:", text)
print("Filtered Text:", filtered_text)
```

### 9.7 Viết chương trình Python với thư viện NLTK để tìm tập hợp các từ đồng nghĩa và trái nghĩa của một từ nào đó.

```python
import nltk
from nltk.corpus import wordnet

def get_word_info(word):
    # Tải xuống dữ liệu nếu chưa có
    nltk.download('wordnet')

    # Tìm các synsets của từ
    synsets = wordnet.synsets(word)

    if not synsets:
        return f"No definition found for '{word}'"

    # Lấy định nghĩa và ví dụ từ synset đầu tiên
    definition = synsets[0].definition()
    examples = synsets[0].examples()

    # Tìm tập hợp từ đồng nghĩa và trái nghĩa
    synonyms = set()
    antonyms = set()

    for syn in synsets:
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())

    return definition, examples, synonyms, antonyms

# Từ cần tra cứu
word = "computer"

definition, examples, synonyms, antonyms = get_word_info(word)

print(f"Word: {word}")
print(f"Definition: {definition}")
print("Examples:")
for example in examples:
    print(f"- {example}")
print("\nSynonyms:", ", ".join(synonyms))
print("Antonyms:", ", ".join(antonyms))
```

### 9.9 Viết chương trình Python với thư viện NLTK để so sánh sự giống nhau của hai danh từ đã cho.

```python
import nltk
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag

def get_word_info(word):
    # Tải xuống dữ liệu nếu chưa có
    nltk.download('wordnet')
    nltk.download('averaged_perceptron_tagger')
    nltk.download('punkt')

    # Tìm các synsets của từ
    synsets = wordnet.synsets(word)

    if not synsets:
        return f"No definition found for '{word}'"

    # Lấy định nghĩa và ví dụ từ synset đầu tiên
    definition = synsets[0].definition()
    examples = synsets[0].examples()

    # Tìm tập hợp từ đồng nghĩa và trái nghĩa
    synonyms = set()
    antonyms = set()

    for syn in synsets:
        for lemma in syn.lemmas():
            synonyms.add(lemma.name())
            if lemma.antonyms():
                antonyms.add(lemma.antonyms()[0].name())

    return definition, examples, synonyms, antonyms

def get_pos_tags(sentence):
    # Tokenize và gán nhãn POS
    words = word_tokenize(sentence)
    pos_tags = pos_tag(words)

    return pos_tags

def compare_noun_similarity(noun1, noun2):
    nltk.download('omw-1.4')
    syn1 = wordnet.synsets(noun1, pos=wordnet.NOUN)
    syn2 = wordnet.synsets(noun2, pos=wordnet.NOUN)

    if not syn1 or not syn2:
        return f"No similarity found between '{noun1}' and '{noun2}'"

    # Lấy synset đầu tiên và tính độ tương đồng
    similarity = syn1[0].wup_similarity(syn2[0])
    return similarity if similarity is not None else 0.0

# Từ cần tra cứu
word = "computer"
definition, examples, synonyms, antonyms = get_word_info(word)

print(f"Word: {word}")
print(f"Definition: {definition}")
print("Examples:")
for example in examples:
    print(f"- {example}")
print("\nSynonyms:", ", ".join(synonyms))
print("Antonyms:", ", ".join(antonyms))

# Câu ví dụ để gán nhãn POS
sentence = "The quick brown fox jumps over the lazy dog."
print("\nPOS Tags:")
print(get_pos_tags(sentence))

# So sánh độ tương đồng giữa hai danh từ
noun1 = "computer"
noun2 = "laptop"
similarity_score = compare_noun_similarity(noun1, noun2)
print(f"\nSimilarity between '{noun1}' and '{noun2}': {similarity_score}")
```
