# NUMPY

### 1.1: Tạo mảng với NumPy

```python

import numpy as np

# TẠO MẢNG 1 CHIỀU
mang_1_chieu = np.array([1,3, 5, 7, 9])
print("Mảng 1 CHIỀU:", mang_1_chieu)

# TẠO MẢNG 2 CHIỀU
mang_2_chieu = np.array([[1,3,5] , [7,9,11]])
print("Mảng 2 chiều", mang_2_chieu)
# TẠO MẠNG NGẪU NHIÊN
mang_ngau_nhien = np.random.rand(3,3)
print("Mảng số ngẫu nhiên:", mang_ngau_nhien)
```

### **1.2: Các thao tác cơ bản của Numpy**

```python
mang_1_chieu = np.array([1,3,5,7,9])
# Cộng trừ nhân chia với một số
print("Cộng thêm 10:", mang_1_chieu + 10)
print("Nhân với 5:", mang_1_chieu * 5)
# Tính tổng , Trung bình , lớn nhất , nhỏ nhất
print("Tổng:", mang_1_chieu.sum())
print("Trung bình:", mang_1_chieu.mean())
print("Giá trị lớn nhất:", mang_1_chieu.max())
print("Giá trị nhỏ nhất:", mang_1_chieu.min())
# lấy phần tử bất kỳ trong mảng
print("Phần tử đầu tiên:", mang_1_chieu[0])
```

## **Pandas**

```python
import pandas as pd

data = {
    "Tên": ["Phát", "Ý", "Thịnh"],
    "Tuổi": [21, 21, 35],
    "Điểm": [85, 90, 95]
}
df =pd.DataFrame(data)
print(df)
```

```python
# In ra cột tuổi
print("Cột tuổi:\n", df["Tuổi"])
print("Hàng đầu tiên:\n", df.iloc[0])

# In ra dữ liệu có điểm lớn hơn 80
filtered_df = df[df["Điểm"] > 80]

print("Học sinh có điểm > 80:\n", filtered_df)

# Tính điểm trung bình của lớp học
print("Tuổi trung bình:", df["Tuổi"].mean())
```

## **Matplotlib**

```python
import matplotlib.pyplot as plt

# Vẽ biểu đồ đường
x = [1, 2, 3, 4, 5]
y = [20, 5, 15, 25, 2]

plt.plot(x, y, label="Biểu đồ đường", color="blue", marker="o")
plt.title("Biểu đồ đường cơ bản")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.legend()
plt.show()

```

```python
# Vẽ biểu đồ cột
categories = ["Thịnh","Trường","Ý","Phát"]
values = [6, 9, 5, 8]

plt.bar(categories, values, color="skyblue")
plt.title("Biểu đồ cột")
plt.xlabel("Danh mục")
plt.ylabel("Giá trị")
plt.show()
```

### **Biểu đồ phân tán**

```python
# Vẽ biểu đồ phân tán
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.rand(100) * 2000

plt.scatter(x, y, s=sizes, alpha=0.5, c='pink')
plt.title("Biểu đồ phân tán")
plt.xlabel("Trục X")
plt.ylabel("Trục Y")
plt.show()
```
